```python
import pandas as pd

# Load dataset (Assume you have the 'HateSpeech_Kenya.csv' file)
data = pd.read_csv('/content/HateSpeech_Kenya.csv')
print(data.head())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Summary statistics
print(data.describe())
```

```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Text Preprocessing
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Tokenize and lower the case
    tokens = nltk.word_tokenize(text.lower())
    # Remove stop words
    tokens = [word for word in tokens if word not in stop_words and word.isalpha()]
    return ' '.join(tokens)

# Apply preprocessing to the 'Tweet' column
data['cleaned_tweet'] = data['Tweet'].apply(preprocess_text)

# Vectorization using TF-IDF
tfidf = TfidfVectorizer(max_features=1000)  # You can adjust the max_features based on your needs
X = tfidf.fit_transform(data['cleaned_tweet'])

# View the first few rows of transformed data
print(X.toarray()[:5])
```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns
numerical_cols = ['hate_speech', 'offensive_language', 'neither']

# Initialize scaler
scaler = StandardScaler()

# Scale numerical features
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Display scaled features
print(data[numerical_cols].head())
```

```python
# Define features (X) and target (y)
X = tfidf.fit_transform(data['cleaned_tweet'])  # Text data
y = data['Class']  # Assuming this is the label for classification


```

```python
#split the data for training

x=data.drop('cleaned_tweet', axis=1)
y=data['Class']
```

```python
from sklearn.model_selection import train_test_split

#Split data into training and testing sets
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)

print(f"Train set size: {x_train.shape}")
print(f"Train set size: {x_test.shape}")
```
